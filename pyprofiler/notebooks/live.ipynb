```python
#map to OMA HOGs
def grabHog(ID):
    try:
        return p.db_obj.entry_by_entry_nr(p.db_obj.id_resolver.resolve(ID))[4].decode()
    except:
        return np.nan
```

```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
for q in qs:
    qHOG = grabHog(q)
    qhogs.append(qHOG)
    start = time.time()
    
    res= p.hog_query( hog_id=str(qHOG) , k = 20 )
    
    results += res
    print(str( time.time()-start )  + 'seconds elapsed' )
    hashes = p.pull_hashes( set(res+ [qHOG]) )
    
    jkern = [ [ h1.jaccard(h2) if i<j else 0 for j,h2 in enumerate(hashes.values()) ] for i,h1 in enumerate(hashes.values())]
    jkern = np.asarray(jkern)
    jkern+=jkern.T
    np.fill_diagonal(jkern , 1 )
    g = sns.clustermap( jkern , xticklabels=hashes.keys()  , yticklabels=hashes.keys(), figsize = (10,10) )
    print(res)
    
    goares = goa.run_GOEA_onresults( res  , p.db_obj, study )
    #output the results
    goea_results_sig = [r for r in goares[0] if r.p_fdr_bh < 0.1]
    study.wr_txt( q+"_zp.txt", goea_results_sig)
    plt.show()
    
hashes = p.pull_hashes( set(results+qhogs) )
print(hashes['HOG:0363809.1b.1a.1c.1a'].digest())
jkern = [ [ h1.jaccard(h2) if i<j else 0 for j,h2 in enumerate(hashes.values()) ] for i,h1 in enumerate(hashes.values())]
jkern = np.asarray(jkern)
jkern+=jkern.T
np.fill_diagonal(jkern , 1 )
g = sns.clustermap( jkern , xticklabels=hashes.keys()  , yticklabels=hashes.keys(), figsize = (10,10) )
```
