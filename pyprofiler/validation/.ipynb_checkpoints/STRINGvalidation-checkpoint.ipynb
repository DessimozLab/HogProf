```python
#set up redis DB with string data

r1 = redis.StrictRedis(host='localhost', port=6379, db=1)

#sort the IDs alphanumerically.
#protein1 protein2 neighborhood fusion cooccurence coexpression 
#experimental database textmining combined_score
#394.NGR_c00010 394.NGR_c33930 0 0 165 0 0 0 145 255

#37200000

refs = ['neighborhood', 'fusion', 'cooccurence', 
        'coexpression', 'experimental', 'database', 'textmining', 'combined_score']
startline = 37200000

with open('/home/cactuskid/mountptLSF/stringdata/protein.links.detailed.v10.5.txt', 'r')as stringAll:
    for i, line in enumerate(stringAll):
        try:
            if i > startline:
                words = line.split()
                rowdict = {}
                rowdict = { refs[i]:entry for i,entry in enumerate(words[2:]) }
                IDS = [words[0], words[1]]
                interaction = IDS[0]+'-'+IDS[1]
                r1.set(interaction, json.dumps(rowdict))
            if i % 100000 == 0 :
                print(i)
            if i % 1000000 == 0 :
                gc.collect()
        except:
            print(i)
```
