```python
targets  = pd.read_table( './out_map_by_hog.tsv')
print(len(targets.cafa_id.unique()))
targets['species'] = targets.oma_id.map(lambda x : x[0:5])
print(targets.head())
print(targets.species.unique())
```

```python
#write out each set by taxcode
'''9606 	# "Homo sapiens"
10090 	# "Mus musculus[All Names]"
10116 	# "Rattus norvegicus"        
3702 	# "Arabidopsis thaliana[All Names]"
83333 	# "Escherichia coli K-12[all names]" P
7227 	# "Drosophila melanogaster[All Names]" 
287 	# "Pseudomonas aeruginosa[All Names]" P
559292 	# "Saccharomyces cerevisiae ATCC 204508"
284812 	# "Schizosaccharomyces pombe ATCC 24843"
7955 	# "Danio rerio[All Names]"
44689 	# "Dictyostelium discoideum[All Names]"
243273 	# "Mycoplasma genitalium ATCC 33530" P
6239 	# "Caenorhabditis elegans[All Names]"
226900 	# "Bacillus cereus ATCC 14579" P
4577 	# "Zea Mays [All names]"
9823 	# "Sus scrofa" 
99287   # Salmonella typhymurium ATCC 700720'''
```

```python
specmap = { 'DICDI': 44689 , 'MAIZE': 4577 , 'ARATH':3702 ,  'SCHPO':284812 , 'YEAST':559292, 'CAEEL':6239 , 'DROME':7227, 'DANRE':7955 , 'PIGXX':9823,
 'MOUSE':10090, 'RATNO':10116 , 'HUMAN':9606 , 'MYCGE':243273, 'BACCR':226900, 'PSEAE':287 ,  'SALTY':99287 , 'ECOLI':83333 }
targets['taxcode'] = targets.species.map(specmap)
print(targets.head())
```

```python
header= '''AUTHOR profiler_dessimozlab
MODEL 1
KEYWORDS phylogeny , protein interactions.
'''

footer = 'END'

filenameproto = 'dessimozlab_1_taxonID_go.txt'

for code in targets.taxcode.unique():
    with open( './profiler_dessimozlab/' + filenameproto.replace('taxonID' , str(code) ) , 'w' ) as outfile:
        print(code)
        outfile.write( header)
        sub = targets[targets.taxcode == code]
        subpreds = preds[ preds.index.isin(sub.cafa_id)]
        
        subpreds.to_csv( outfile , sep = '\t' , header = None   )
        print(subpreds.head())
        outfile.write(footer)
```

```python
header= '''AUTHOR profiler_dessimozlab
MODEL 2
KEYWORDS phylogeny , protein interactions, machine learning.
'''

footer = 'END'

filenameproto = 'dessimozlab_2_taxonID_go.txt'

for code in targets.taxcode.unique():
    with open( './profiler_dessimozlab/' + filenameproto.replace('taxonID' , str(code) ) , 'w' ) as outfile:
        print(code)
        
        outfile.write( header)
        sub = targets[targets.taxcode == code]
        subpreds = preds2[ preds2.index.isin(sub.cafa_id)]
        subpreds.to_csv( outfile , sep = '\t' , header = None  )
        print(subpreds.head())
        outfile.write(footer)
```
